The dataset contains the following columns:

- InvoiceNo: Invoice number of the transaction
- StockCode: Unique code of the product
- Description: Description of the product
- Quantity: Quantity of the product in the transaction
- InvoiceDate: Date and time of the transaction
- UnitPrice: Unit price of the product
- CustomerID: Unique identifier of the customer
- Country: Country where the transaction occurred

# Task 3: Cleaning the data:

# Cancelled/ Returned orders:

The description of the dataset suggests that there were 80995 orders that were returned or cancelled. We can check whether the negative sign means an order cancelled or returned by doing the following:

```python
cancelled_orders = dataset[dataset['InvoiceNo'].astype(str).str.contains('C')]
cancelled_orders.head()
```

```python
cancelled_orders.shape
```

```python
cancelled_orders[cancelled_orders['Quantity']>0]
```

As expected, when the quantity is positive we can see that there are no cancelled orders thus, we can conclude that negative sign symbolises cancelled/ returned order. Also, we can see that there were 8872 cancelled orders in that time period.

```python
duplicate_rows_dataset = dataset[dataset.duplicated()]
print("number of duplicate rows: ", duplicate_rows_dataset.shape)
```

There are 5268 duplicate rows. Thus we will remove 5268 rows out of the total 541909 rows.

```python
df = dataset.drop_duplicates()
dataset.head(5)
```

```python
dataset.count()
```

We can notice that **Description** and **Customer ID** column has some missing values as they only have 540455 and 406829 values respectively. We need to consider them during exploration.

```python
dataset = dataset.drop_duplicates()
dataset.count()
```

```python
print(dataset.isnull().sum())
```

```python
dataset = dataset.dropna()    # Dropping the missing values.
dataset.count()
```

```python
print(dataset.isnull().sum())
```

Now we have dropped all the missing values from the dataset. We can see after dropping the null values from the datset, all the variables in the datset has equal number of entries.

```python
#remove cancelled orders
dataset['InvoiceNo'] = dataset['InvoiceNo'].astype('str')
dataset = dataset[~dataset['InvoiceNo'].str.contains('C')]
```

```python
dataset.shape
```

#Visualization:

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.5, 8)
ax=sns.barplot(x='Country', y='total_cost',data=dataset,estimator=max,ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=47, ha="right")
plt.show()
```

As expected since the most number of customers and most number of quantity is ordered from United Kingdom, the online retail attracts the most revenue from the respective country itself.
